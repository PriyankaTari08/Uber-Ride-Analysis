```python
df['pickup_time'] = pd.to_datetime(df['pickup_time'])
df['drop_time'] = pd.to_datetime(df['drop_time'])

df[['pickup_time', 'drop_time']].dtypes
```

```python
#Why convert object to datetime?
#Right now, Python sees those columns as plain text, so it can't answer questions like:
#What is the busiest hour? Which day has the most rides? How many rides happened each month? What is the average ride duration?
#By converting them to datetime, you unlock all of those analyses.
```

**Phase 2: Feature Engineering**

Now we're going to create new columns from pickup_time. This is one of the most important steps because it makes time-based analysis much easier.

```python
df['pickup_hour'] = df['pickup_time'].dt.hour
df['pickup_day'] = df['pickup_time'].dt.day
df['pickup_month'] = df['pickup_time'].dt.month
df['pickup_weekday'] = df['pickup_time'].dt.day_name()
```

```python
#To check the status
df['status'].value_counts()
```

```python
# Calculate total rides per city
total_rides_per_city = df['city'].value_counts()

# Calculate cancelled rides (including 'No-Show') per city
cancelled_rides_per_city = df[df['status'].isin(['Cancelled', 'No-Show'])]['city'].value_counts()

# Calculate cancellation rate
cancellation_rate_per_city = (cancelled_rides_per_city / total_rides_per_city) * 100

# Display the cancellation rates sorted in descending order
display(cancellation_rate_per_city.sort_values(ascending=False))
```

```python
# Get the city with the highest cancellation rate
city_highest_cancellation_rate = cancellation_rate_per_city.idxmax()

print(f"The city with the highest cancellation rate is: {city_highest_cancellation_rate}")
```

### City Performance Summary

```python
# Calculate total revenue per city
total_revenue_per_city = df.groupby('city')['fare_amount'].sum()

# Calculate average fare per city
average_fare_per_city = df.groupby('city')['fare_amount'].mean()

# Calculate average duration per city
average_duration_per_city = df.groupby('city')['duration'].mean()

# Combine all metrics into a single DataFrame
city_performance_summary = pd.DataFrame({
    'Total Rides': total_rides_per_city,
    'Total Revenue': total_revenue_per_city,
    'Average Fare': average_fare_per_city,
    'Cancellation Rate (%)': cancellation_rate_per_city,
    'Average Ride Duration': average_duration_per_city
})

# Display the summary, sorted by Total Revenue for better insight
display(city_performance_summary.sort_values(by='Total Revenue', ascending=False))
```

**Why I'm emphasizing this table**

Imagine an interviewer asks:

"Can you compare the performance of different cities?"

--Instead of opening five different charts, you can point to one summary table and explain:

Which city has the most rides.
Which city earns the most revenue.
Which city has the highest average fare.
Which city has the longest average distance.
Which city has the longest average ride duration.

| City                         | Observation                                              |
| ---------------------------- | -------------------------------------------------------- |
| **Boston**                   | Highest rides, highest revenue, strong overall performer |
| **Seattle**                  | Highest average fare and longest average ride duration   |
| **Los Angeles**              | Lowest average fare and highest cancellation rate        |
| **Chicago**                  | Shortest average ride duration                           |
| **San Francisco & New York** | Balanced performance across all metrics                  |


**Recommendation**

Boston is the strongest-performing market. Uber should ensure sufficient driver availability to maintain service quality and support continued growth.

Seattle appears to have higher-value rides on average. Uber could further analyze the factors contributing to these longer or higher-value trips.

Los Angeles should be investigated to understand why cancellations are comparatively higher and why average fares are lower.
